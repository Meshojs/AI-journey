```python
from matplotlib import pyplot as plt 
import numpy as np
import matplotlib as mpl

# mpl.rcParams("text.color") = "black"


plt.figure(figsize=(20 , 10))
plt.style.use("fivethirtyeight")
age = [20 , 21 , 22 , 23]
salary_dev =[2000 , 3200 , 4100 , 6000]
python_dev = [2100 ,3300 , 8900 , 10000]

ns = np.array(salary_dev  )
pd = np.array(python_dev )

plt.plot(age ,salary_dev , label="salary of dev")
plt.plot(age , python_dev ,label ="salary of python dev")
plt.fill_between(age,ns , alpha=0.1)
plt.fill_between(age,pd ,alpha=0.1)
plt.legend()


plt.show()
```
